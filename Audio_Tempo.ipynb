```python
from google.colab import files # Import the files object from google.colab

def analyze_tempo(file_path):
    """
    Analyzes the tempo of a WAV file.

    Args:
        file_path (str): Path to your WAV file

    Returns:
        tuple: (tempo in BPM, beat frame indices)
    """

    y, sr = librosa.load(file_path)

    tempo, beats = librosa.beat.beat_track(y=y, sr=sr)

    beat_times = librosa.frames_to_time(beats, sr=sr)

    return tempo, beats, beat_times

print("Please select your WAV file:")
uploaded = files.upload()

filename = next(iter(uploaded))

tempo, beats, beat_times = analyze_tempo(filename)

print(f"\nEstimated tempo: {float(tempo):.2f} BPM")
print("\nBeat timestamps (in seconds):")
for i, time in enumerate(beat_times):
    print(f"Beat {i+1}: {float(time):.2f}s")
```
